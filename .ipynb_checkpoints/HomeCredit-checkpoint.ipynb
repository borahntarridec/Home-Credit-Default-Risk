```python
def bureau_aggregation(new_df, df):
    AGGREGATIONS = [{'groupby':['SK_ID_CURR'],'select':'SK_ID_BUREAU','agg':'nunique'},
                    {'groupby':['SK_ID_CURR'],'select':'CREDIT_TYPE','agg':'nunique'},
                    {'groupby':['SK_ID_CURR'],'select':'DAYS_CREDIT','agg':'mean'},
                    {'groupby':['SK_ID_CURR'],'select':'DAYS_CREDIT','agg':'max'},
                    {'groupby':['SK_ID_CURR'],'select':'CREDIT_DAY_OVERDUE','agg':'mean'},
                    {'groupby':['SK_ID_CURR'],'select':'CREDIT_DAY_OVERDUE','agg':'max'},      
                   ]
    for agg in AGGREGATIONS:
        # features to select
        all_features = list(set(agg['groupby'] + [agg['select']]))
        # variables name
        new_feature = '{0}_{1}'.format(agg['select'], agg['agg'])
        print(new_feature)
        # perform the groupby
        gp = df[all_features] \
                .groupby(agg['groupby']) \
                [agg['select']] \
                .agg(agg['agg']) \
                .reset_index()  \
                .rename(index=str, columns={agg['select']:new_feature})
        # Merge
        new_df = new_df.merge(gp,on=agg['groupby'],how='left')
    return new_df
```

```python
ml_bureau = bureau_aggregation(ml_bureau, bureau)
```

```python
def encode_categorie(new_df, df, to_encode):
    
    """ to_encode : a list of columns name to encode """
    
    for col in to_encode:
        le=LabelEncoder()
        le.fit(df[col])
        new_df[col] = le.transform(df[col])
    return new_df
```

```python
ml_bureau = encode_categorie(ml_bureau, bureau, ['CREDIT_ACTIVE','CREDIT_CURRENCY','CREDIT_TYPE'])
```

```python
def bureau_balance_aggregation(new_df, df):
    AGGREGATIONS = [{'groupby':['SK_ID_BUREAU'],'select':'MONTHS_BALANCE','agg':'min'},
                    {'groupby':['SK_ID_BUREAU'],'select':'MONTHS_BALANCE','agg':'max'},
                   ]
    for agg in AGGREGATIONS:
        # features to select
        all_features = list(set(agg['groupby'] + [agg['select']]))
        # variables name
        new_feature = '{0}_{1}'.format(agg['select'], agg['agg'])
        print(new_feature)
        # perform the groupby
        gp = df[all_features] \
                .groupby(agg['groupby']) \
                [agg['select']] \
                .agg(agg['agg']) \
                .reset_index()  \
                .rename(index=str, columns={agg['select']:new_feature})
        # Merge
        new_df = new_df.merge(gp,on=agg['groupby'],how='left')
    return new_df
```

```python
def compute_share(df):
    df['INCOME_CREDIT_LEVEL'] = df['AMT_CREDIT'] / df['AMT_INCOME_TOTAL']
    df['ANNUITY_INCOME_SHARE'] = df['AMT_ANNUITY'] / df['AMT_INCOME_TOTAL']
    df['GOODS_PRICE_CREDIT_SHARE'] = df['AMT_GOODS_PRICE'] / df['AMT_CREDIT']
    df['ANNUITY_CREDIT_SHARE'] = df['AMT_ANNUITY'] / df['AMT_CREDIT']
    return df
app_train = compute_share(app_train)
```

```python
def fillmissing_value(df):
    # int, fill nan with the mean
    for col in df[df.columns[df.isnull().any()]].columns:
        if df[col].dtypes != 'object':
            df[col].fillna(df[col].mean(), inplace=True)
    # string, fill na by unknown
    df.fillna('unknown', inplace=True)
    return df

app_train = fillmissing_value(app_train)
```

```python
app_train = encode_categorie(app_train, app_train, app_train.select_dtypes(include='object').columns.tolist())
```
